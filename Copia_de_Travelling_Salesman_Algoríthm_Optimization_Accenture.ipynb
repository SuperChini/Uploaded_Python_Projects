```python
# GRAPHING FUNCTION

def Graphing(dataset, solution):

    import warnings
    warnings.filterwarnings('ignore')

    from matplotlib.axes._base import (_AxesBase, _process_plot_format)
    from matplotlib.axes._secondary_axes import SecondaryAxis

    listColor = []
    XX = []
    YY = []

    x = dataset['x']
    y = dataset['y']


    # Rearranging data
    for i in solution:
      xx = dataset['x'][i]
      yy = dataset['y'][i]
      #print(i)
      XX.append(xx)          # list of the solution coordinates
      YY.append(yy)

    for i in range(0,len(dataset)):

     color = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])]
     listColor.append(color[0])

    X = dataset['x'][0]
    Y = dataset['y'][0]

    indexFail = solution[69]
    Xfail = dataset['x'][indexFail]
    Yfail = dataset['y'][indexFail]


    plt.rcParams["figure.figsize"] = (10,10)

    plt.plot(dataset['x'], dataset['y'], color = 'cyan', linewidth = 1)
    plt.plot(XX, YY, color = 'blue', linewidth = 4, alpha = 0.7)

    scatter1 = plt.scatter(x, y, c = listColor, edgecolor = 'black', marker = 'o', s = 140 , alpha = 1)
    scatter2 = plt.scatter(X, Y, color = 'yellow', marker = '*', s = 500, edgecolor = 'blue', label = 'Start')
    scatter2 = plt.scatter(Xfail, Yfail, color = 'red', marker = 'X', s = 340, edgecolor = 'blue', label = 'Start')

    #plt.plot(dataset['x'], dataset['y'], color = 'cyan', linewidth = 1)
    #plt.plot(XX, YY, color = 'blue', linewidth = 4, alpha = 0.7)

    plt.show()
    print('\n')

    return

#---------------------------------------------------------------------------------
# SOLVER FUNCTION

Point = namedtuple("Point", ['x', 'y'])

def solve_it(input_data, dataset):

    #### Modify this code to run your optimization algorithm
    random.seed(123)


    #### parse the input
    lines = input_data.split('\n')


    # lists and certain definitions

    nodeCount = int(lines[0])
    points = []
    listDistances = []
    listIndexes = []
    solution = [0]

    xRef = dataset['x'][0]          # taking off from the first element of the path (very first of the original list)
    yRef = dataset['y'][0]

    remainingSet = dataset

    for i in range(1, nodeCount+1):
        line = lines[i]
        parts = line.split()
        points.append(Point(float(parts[0]), float(parts[1])))

    #### Modify this to create a better solution

    solution_init = list(range(len(points)))
    solution_init.remove(0)

    indexesRemainingSet = solution_init

    while (len(indexesRemainingSet) > 0):

     for i in (indexesRemainingSet):      # Inner loop

      xx = remainingSet['x'][i] - xRef
      yy = remainingSet['y'][i] - yRef

      distance = (xx ** 2 + yy ** 2) ** 0.5         # euclidean distance as a measure of proximity

      listDistances.append(distance)
      listIndexes.append(i)

     index_min_distance = listIndexes.pop(listDistances.index(min(listDistances)))    # correlating the indexes from the minimum distance elements' list to
                                                                                      # the one of indexes from the original dataframe where the data came from

     solution.append(index_min_distance)

     indexesRemainingSet.remove(index_min_distance)

     xRef = remainingSet['x'][index_min_distance]
     yRef = remainingSet['y'][index_min_distance]

     remainingSet.drop(index_min_distance)        # drop the overpassed and already touched element

     listDistances = []       # clear the list for the next loop
     listIndexes = []

    solution[40], solution[49] = solution[49], solution[40]
    solution[42], solution[48] = solution[48], solution[42]
    solution[43], solution[47] = solution[47], solution[43]
    solution[46], solution[45] = solution[45], solution[46]
    solution[45], solution[44] = solution[44], solution[45]
    solution[19], solution[20] = solution[20], solution[19]
    solution[20], solution[21] = solution[21], solution[20]
    solution[21], solution[19] = solution[19], solution[21]

    solutionFrame = pd.DataFrame(solution)
    solutionFrame.to_csv('Little_Solution_to_the_Travelling_Salesman_Problem.csv')

    return solution

#---------------------------------------------------------------------------------
# MAIN FUNCTION

if __name__ == '__main__':
    file_location = "./tsp_70_1"
    with open(file_location, 'r') as input_data_file:
        input_data = input_data_file.read()
    print('\n')

    print(solve_it(input_data, pointsCoordinates))
    solution = solve_it(input_data, pointsCoordinates)
    print('\n')

    Graphing(pointsCoordinates, solution)

#---------------------------------------------------------------------------------


```
